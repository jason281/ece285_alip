```python
import os
import PIL.ImageOps
from PIL import Image
import numpy as np
from tqdm import tqdm as tqdm
```

```python
for class_ in tqdm(os.listdir('img')):
    if not os.path.exists('background/'+class_):
        os.mkdir('background/'+class_)
    if not os.path.exists('foreground/'+class_):
        os.mkdir('foreground/'+class_)
    
    for filename in tqdm(os.listdir(os.path.join('img', class_))):
        filename = filename.strip()
        img_path = os.path.join('img', class_, filename)
        mask_path = os.path.join('mask', class_, filename)
        if not os.path.exists(mask_path):
            f, ext = os.path.splitext(filename)
            filename = f + '.png'
            for ext in ['.png', '.jpg', '.jpeg']:
                mask_path = os.path.join('mask', class_, f+ext)
                if os.path.exists(mask_path):
                    break
        
        bg_path = os.path.join('background', class_, filename)
        fg_path = os.path.join('foreground', class_, filename)
        if os.path.exists(bg_path) and os.path.exists(fg_path):
            continue
        
        img, mask = Image.open(img_path), Image.open(mask_path)
        background, foreground = Image.composite(0, img, mask), Image.composite(0, img, PIL.ImageOps.invert(mask))
        
        background.convert('RGB').save(bg_path)
        foreground.convert('RGB').save(fg_path)
```
